```python
battery_drain_per_mine = 15
profit_target = 100

def mine_asteroid(size, scan_clearance=False):
    # No input() inside here! Use the parameters provided.
    if scan_clearance == True:
        minerals = size * 2
    else:
        if size > 5:
            minerals = size * 1.5
        else:
            minerals = 0
    return minerals

def process_and_sell(minerals):
    if minerals > 20:
        profit = minerals * 3
    else:
        profit = minerals * 2
    return profit

total_profit = 0
drone_battery = 100

while True:
    print(f"\n--- Status: Profit = {total_profit}/{profit_target} | Battery = {drone_battery}% ---")
    choice = input("Enter command (mine/quit): ").strip().lower()

    if choice == "quit":
        print("Exiting simulator. Goodbye!")
        break
    elif choice != "mine":
        # If they type anything else, pass and restart the loop
        print("Unknown system command.")
        pass
    else:
        # Gather inputs in the main execution block
        asteroid_size = int(input("Enter asteroid size (1-10): "))
        clearance_input = input("Clearance code? (True/False): ")

        # Proper boolean check
        clearance = (clearance_input == "True")

        # Deduct battery
        drone_battery -= battery_drain_per_mine

        # CONNECTING THE PIPELINE:
        # 1. Mine the asteroid and store returned minerals
        minerals_gathered = mine_asteroid(asteroid_size, clearance)

        # 2. Pass minerals to refinery and store returned profit
        round_profit = process_and_sell(minerals_gathered)

        # 3. Add to global tracker
        total_profit += round_profit
        print(f"Mined {minerals_gathered} minerals. Earned {round_profit} credits!")

    # GAME OVER CHECKS
    if total_profit >= profit_target:
        print("\n🏆 Victory! Profit target met! You are a master prospector.")
        break
    if drone_battery <= 0:
        print("\n💀 System Crash! Battery depleted in deep space.")
        break
```

```python
####The "PhonePe" Security & Payment Matrix
class PhonePeWallet:
    def __init__(self, owner, opening_balance):
        self.owner = owner
        # 1. Encapsulate the balance attribute to make it private
        self.__balance = opening_balance

    def get_balance(self):
        # Getter method to safely view the private balance
        return self.__balance


# 2. Inherit UPIPayment from PhonePeWallet parent class
class UPIPayment(PhonePeWallet):

    # 3. Polymorphism: Implement custom logic for process_payment
    def process_payment(self, amount):
        convenience_fee = 2
        total_deduction = amount + convenience_fee

        # Accessing parent properties safely
        if self.get_balance() >= total_deduction:
            # Updating the private balance
            self._PhonePeWallet__balance -= total_deduction
            print(f"UPI Payment of ₹{amount} (+₹{convenience_fee} Fee) Successful!")
            print(f"Remaining Wallet Balance: ₹{self.get_balance()}")
        else:
            print("Transaction Failed: Insufficient Funds!")

# --- Execution Queue ---
user_wallet = UPIPayment("Rahul", 500)

# Process a transaction
user_wallet.process_payment(100)
```

```python
###The "Swiggy Delivery" Token Dispenser
def swiggy_token_generator(limit):
    token_number = 1
    while token_number <= limit:
        token_string = f"SWIGGY-ID-{token_number}"
        # 1. Yield the current token string and pause execution
        yield token_string
        token_number += 1

# --- Operational Execution ---

# Spawning the lazy generator object pointer
token_stream = swiggy_token_generator(3)

# 2. Use the next tracker to pull tokens out one by one manually
print("--- Kitchen Order Dispatch Screen ---")
print(f"Dispensing: {next(token_stream)}")
print(f"Dispensing: {next(token_stream)}")
print(f"Dispensing: {next(token_stream)}")
```

```python
##The "Zomato Blinkit" Infinite Supply Chain Streamer
def blinkit_item_streamer(items_list):
  for i in items_list:
    if i["stock"]==0:
      yield f"RESTOCK ALERT: {i['name']} is out of stock!"
inventory = [
    {"name": "Amul Butter", "stock": 5},
    {"name": "Maggi", "stock": 0},
    {"name": "Thums Up", "stock": 0},
    {"name": "Bread", "stock": 12}
]
alert_generator = blinkit_item_streamer(inventory)
alert_iterator=iter(alert_generator)
print(next(alert_iterator))
print(next(alert_iterator))
```

```python
from collections import Counter, defaultdict, deque

# --- TASK 1: FREQUENCY ANALYTICS (Counter) ---
swipe_ins = ["Rajiv Chowk", "Hauz Khas", "Rajiv Chowk", "Kashmere Gate", "Hauz Khas", "Rajiv Chowk"]

# Tally the station entries automatically
station_footfall = Counter(swipe_ins)

# Extract the single busiest station
busiest_station = station_footfall.most_common(1)
print(f"Busiest Station: {busiest_station}")


# --- TASK 2: DYNAMIC ROUTING (defaultdict) ---
# Initialize defaultdict where every new key defaults to an empty list
metro_lines = defaultdict(list)

# Directly append stations to lines without any if-else checks
metro_lines["Yellow Line"].append("Rajiv Chowk")
metro_lines["Yellow Line"].append("Kashmere Gate")
metro_lines["Blue Line"].append("Noida Sec 62")

print(f"Metro Network: {dict(metro_lines)}")


# --- TASK 3: ROLLING BUFFER QUEUE (deque) ---
# Initialize a deque with a strict maximum capacity of 3
recent_alerts = deque(["Alert A", "Alert B"], maxlen=3)

# Add standard alert to the right end
recent_alerts.append("Alert C")

# Add high-priority alert to the right end (forces "Alert A" to drop off left edge)
recent_alerts.append("EMERGENCY ALERT")

print(f"Recent Alerts: {list(recent_alerts)}")
```
